```python
#Importar librerías
import math

#Recepción de números enteros a transformar
lista_numeros_ent=[]
cantidad_ent=input("Ingrese la cantidad de numeros enteros que desea transformar: ")
for i in range(int(cantidad_ent)):
    numero_ent=input("Ingrese un numero entero: ")
    lista_numeros_ent.append(int(numero_ent))


#Recepción de números decimales a transformar
lista_numeros_dec=[]
cantidad_dec=input("Ingrese la cantidad de numeros decimales que desea transformar: ")
for i in range(int(cantidad_dec)):
    numero_dec=input("Ingrese un numero decimales: ")
    lista_numeros_dec.append(numero_dec)

#########################################################
#######################################################   



def entero_binario(numero):
    """
    :param numero: str 
    :return: caracter_entera: str
    caracter_decimal : str
    """
    parte_entera = numero
    caracter_entera = ""

    while parte_entera >= 1 :
        caracter_entera += str(parte_entera % 2)
        parte_entera = parte_entera // 2
    caracter_entera = caracter_entera [::-1]

            
    transformacion = caracter_entera
    
    return transformacion

#########################################################
#######################################################

def decimal_binario(numero):
    """
    :param numero: str 
    :return: caracter_entera: str
    caracter_decimal : str
    """
    nuevo_numero = numero.split(".")
    parte_entera = abs (int(nuevo_numero[0]))
    parte_decimal = float("0." + str(nuevo_numero[1]))
    caracter_entera = ""
    caracter_decimal = ""

    while parte_entera >= 1 :
        caracter_entera += str(parte_entera % 2)
        parte_entera = parte_entera // 2
    caracter_entera = caracter_entera [::-1]

    while True:
        respuesta = parte_decimal*2
        decimal, entero = math.modf(respuesta)
        parte_decimal = decimal
        decimal = str(decimal).split(".")
        nuevo_numero = int(decimal[1][0])

        if (entero == 0 and nuevo_numero == 0) or (entero == 1 and int(decimal[1]) == 0):
            caracter_decimal += str(int(entero))
            break
        else:
            caracter_decimal += str(int(entero))
            
    transformacion = caracter_entera + "." + caracter_decimal
    
    return transformacion
#########################################################
#######################################################

def exponente_y_mantisa(binario):
    """
    :param binario: int
    :return: mantisa : str
    lista_sesgo: str
    """
    binario  = str(binario).split(".")
    exponente = len(binario[0]) -1
    numero_completo = "".join(binario)
    mantisa = numero_completo [1:]
    sesgo = 127 + exponente
    lista_sesgo = ""
    while sesgo >= 1 :
        lista_sesgo += str(sesgo % 2)
        sesgo = sesgo // 2
    lista_sesgo = lista_sesgo[::-1]
    return mantisa ,lista_sesgo

#########################################################
#######################################################

def signo_y_mantiza_normalizada (numero ,mantisa):
    """
    :param numero: str
    :param mantisa: str
    :return: signo ,mantisa
    """
    descomposicion_numero = [ i for i in numero ]
    if descomposicion_numero[0] == "-":
        signo = 1
    elif descomposicion_numero[0] in "123456789" or descomposicion_numero[0] == "+" :
        signo = 0
    n = len(mantisa)
    mantisa_normalizada = mantisa
    while n < 23 :
        mantisa_normalizada += "0"
        n += 1

    return signo , mantisa_normalizada

#########################################################
#######################################################

def ejecucion_total (numero):
    c = decimal_binario(numero)
    mantisa,lista_sesgo = exponente_y_mantisa(c)
    signo , bits = signo_y_mantiza_normalizada (numero ,mantisa)
    print ("signo : ", signo)
    print("exponente : ",lista_sesgo)
    print("mantisa : ",bits)
    print([signo],[lista_sesgo],[bits])
    binario_completo=str(signo)+str(lista_sesgo)+str(bits)
    final=binario_a_decimal(binario_completo) 
    print(final)
    newFile = open("filetest.txt", "wb")
    newFile.write(final.to_bytes(4, byteorder='big')) # ,signed=True
    newFile.close()
    return " "

def entero(cadena):
    #res=[None]*2
    #res = ndarray((2,),byte)
    z = bytes(cadena, 'utf-8')
    print(z)
    #z=list(cadena)
    #enteros=ndarray(z.length(),byte)
    resint = 0
    l = len(z)
    print("longitud " + str(l))
    potencia = l-1;
    for i in range(0,l):
        print("i " + str(i))
        print("z[i] " + str(z[i]))
        # 45 es el ascii del signo -
        if (z[i] == 45 and i==0): 
            potencia-=1
            # Agregar signo, 10000000 00000000, 2^15
            #   Bit 15 contiene signo 0 positivo 1 negativo
            resint+= 0x8000 
        elif (not(z[i]>=48 and z[i]<=57)):
            # 48 es el ascii de 0 y 57 de 9
            raise Exception('No es un número entero')
        else:
            # ** se usa para potenciación
            resint += (z[i]-48)*(10**potencia)
            potencia-=1
    return resint

#Binario a decimal

def binario_a_decimal(binario):
    numero = 0
    pot = len(binario) - 1
    for i in binario:
        aux = int(i) * 2**pot
        pot -= 1
        print(type(aux))
        print(type(numero))
        numero += aux
    numero = str(numero)
    salida=entero(numero)
    print(salida)
    return salida
    


#########################################################
#######################################################

i=0
while i < len(lista_numeros_ent):
    print("el numero ", lista_numeros_ent[i], "transformado a binario es: ")
    nuevo_numero = entero_binario(lista_numeros_ent[i])
    print(nuevo_numero)
    i += 1
    
    
j=0
while j < len(lista_numeros_dec):
    print("el numero ", lista_numeros_dec[j], "transformado a formato IEEE 754 es:")
    ejecucion = ejecucion_total (lista_numeros_dec[j])
    print(ejecucion)
    j += 1
```
